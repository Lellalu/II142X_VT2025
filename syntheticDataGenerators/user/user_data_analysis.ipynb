```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the Swedish users dataset
users_df = pd.read_csv("swedish_users.csv")
users_df

# Popular investment goals - Main Analysis
invest_goals = users_df["invest_goal"].value_counts()
print("Popular Investment Baskets:")
print(invest_goals)
print(f"\nPercentages:")
for goal, count in invest_goals.items():
   pct = (count / len(users_df)) * 100
   print(f"{goal}: {count} users ({pct:.1f}%)")

invest_goals.plot(kind='bar')
plt.title('Popular Investment Baskets by Count')
plt.xlabel('Investment Goal')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```

```python
# Investment basket analysis by demographics
# By Age Group
users_df['age_group'] = pd.cut(users_df['age'], bins=[0, 30, 40, 50, 100], labels=['<30', '30-40', '40-50', '50+'])
age_invest = pd.crosstab(users_df['age_group'], users_df['invest_goal'])
print("\nInvestment Goals by Age Group:")
print(age_invest)

age_invest.plot(kind='bar')
plt.title('Investment Baskets by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# By Location
location_invest = pd.crosstab(users_df['location'], users_df['invest_goal'])
print("\nInvestment Goals by Location:")
print(location_invest)

location_invest.plot(kind='bar')
plt.title('Investment Baskets by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# By Education Level
education_invest = pd.crosstab(users_df['education'], users_df['invest_goal'])
print("\nInvestment Goals by Education:")
print(education_invest)

education_invest.plot(kind='bar')
plt.title('Investment Baskets by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Location distribution
locations = users_df["location"].value_counts()
n_locations = len(locations)
plt.figure(figsize=(8, 4))
colors = plt.cm.Blues(np.linspace(0.9, 0.3, n_locations))
locations.plot(kind='barh', color=colors)
plt.title('Users by Location')
plt.xlabel('Count')
plt.ylabel('Location')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.tight_layout()
plt.show()

print(f"\nUser counts by location:")
for location, count in locations.items():
    print(f"{location}: {count:,}")
```

```python
# Education level distribution
education = users_df["education"].value_counts()
n_education = len(education)
plt.figure(figsize=(8, 4))
colors = plt.cm.Blues(np.linspace(0.9, 0.3, n_education))
education.plot(kind='barh', color=colors)
plt.title('Users by Education Level')
plt.xlabel('Count')
plt.ylabel('Education Level')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.tight_layout()
plt.show()

print(f"\nUser counts by education level:")
for edu_level, count in education.items():
    print(f"{edu_level}: {count:,}")
```

```python
# Gender distribution
gender = users_df["gender"].value_counts()
plt.figure(figsize=(6, 6))
colors = ['#80b1d3','#fb8072']  # Blue and red
gender.plot(kind='pie', autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Users by Gender')
plt.ylabel('')
plt.tight_layout()
plt.show()

print(f"\nUser counts by gender:")
for gender_type, count in gender.items():
    print(f"{gender_type}: {count:,}")
```

```python
# Age distribution
plt.figure(figsize=(8, 4))
plt.hist(users_df['age'], bins=10, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.tight_layout()
plt.show()

print(f"\nAge statistics:")
print(f"Mean age: {users_df['age'].mean():.1f}")
print(f"Median age: {users_df['age'].median():.1f}")
print(f"Age range: {users_df['age'].min()} - {users_df['age'].max()}")
print(f"Standard deviation: {users_df['age'].std():.1f}")
```
